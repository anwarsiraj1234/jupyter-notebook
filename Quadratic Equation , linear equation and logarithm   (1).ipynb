```python
# Quadratic Equation
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function
def quadratic(x, a=11, b=10, c=-90):
    return a*x**2 + b*x + c

# Generate x values
x = np.linspace(-10, 10, 400)

# Generate y values for the quadratic function
y = quadratic(x, a=11, b=10, c=-90)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2 + 2x + 1')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the linear equation
def linear_equation(x):
    return 2 * x + 1

# Generate x values
x_values = np.linspace(-10, 10, 400)
# Generate y values using the linear equation
y_values = linear_equation(x_values)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = 2x + 1', color='blue')

# Add title and labels
plt.title('Graph of y = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
# logairthm 
import sympy as sp

# Define the variables
x = sp.symbols('x')

# Define the logarithmic equation
equation = sp.log(x, 2) - 4

# Solve the equation
solution = sp.solve(equation, x)

# Display the solution
solution
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0.1, 10, 400)  # Avoid x=0 to prevent log(0) which is undefined

# Compute y values for natural logarithm
y = np.log(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = ln(x)', color='blue')

# Add title and labels
plt.title('Logarithmic Graph of y = ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
